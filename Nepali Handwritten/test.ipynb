```python
import cv2
import xml.etree.ElementTree as ET
import os
```

```python
OUTPUT_DIR = 'cropped_images'
DATA_DIR = 'dataset'
```

```python
def crop_image(xml_path, dataset_type='train'):
    tree = ET.parse(xml_path)
    root = tree.getroot()
    image_name = root.find("filename").text
    print(f"Cropping: {image_name}")
    image_path = f'{DATA_DIR}/{dataset_type}/{image_name}'    
    image = cv2.imread(image_path)
    if image is None:
        print(f"Image not found: {image_path}")
        return
    output_path = f'{OUTPUT_DIR}/{dataset_type}'
        
    # Loop through all <object> elements and extract bounding box info
    for i, obj in enumerate(root.findall("object")):
        cropped_path = f"{output_path}/cropped_{image_name.split('.')[0]}_{i}.jpg"
        if os.path.exists(cropped_path):            
            return
        bndbox = obj.find("bndbox")
        xmin = int(bndbox.find("xmin").text)
        ymin = int(bndbox.find("ymin").text)
        xmax = int(bndbox.find("xmax").text)
        ymax = int(bndbox.find("ymax").text)
        h, w, _ = image.shape  # Get image dimensions
        if xmin < 0 or ymin < 0 or xmax > w or ymax > h:
            print(f"⚠️ Skipping invalid bbox: {xmin, ymin, xmax, ymax} (Out of bounds)")
            continue


        # Crop the image
        cropped_img = image[ymin:ymax, xmin:xmax]
        if cropped_img is None or cropped_img.size == 0:
            print(f"Cropped image none: {image_path}")
            return
        # Save cropped image
        cv2.imwrite(cropped_path, cropped_img)
```

```python
# Load and parse the XML file
#xml_path = "dataset/train/891.xml"  # Change this to your XML file
xml_path = f'{DATA_DIR}/train/891.xml'
crop_image(xml_path)
```

```python
dataset_type = 'test'
```

```python
dataset = os.listdir(f'{DATA_DIR}/{dataset_type}')

for file in dataset:
    if file.endswith('.xml'):
        xml_path = f'{DATA_DIR}/{dataset_type}/{file}'
        crop_image(xml_path, dataset_type)
```
